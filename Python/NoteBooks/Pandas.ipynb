```python
import pandas as pd
```

Import Data

```python
sales_data = pd.read_csv('../../Data/Pandas - Ecommerce Purchases.csv')
```

```python
# Check the first indexes
sales_data.head(15)
```

```python
# Quick Count

# Total Columns
print(sales_data.columns)
print(f'Total Columns: {len(sales_data.columns)}',end='\n\n')

# Total Index (rows)
print(sales_data.index)
print(f'Total Index: {len(sales_data.index)}')
```

```python
# Quick Info
sales_data.info()
```

```python
# Purchase price quick stat
print(f'Avg. Purchase Price: {sales_data["Purchase Price"].mean()}')
print(f'Max Purchase Price: {sales_data["Purchase Price"].max()}')
print(f'Min Purchase Price: {sales_data["Purchase Price"].min()}')
```

```python
# Conditional Count
sales_data[sales_data['Job'] == 'Dancer']['Job'].count()
```

```python
# Pivot/Value Count, top 10
sales_data['Job'].value_counts().head(10)
```

```python
# Filtering, show only certain column
sales_data[sales_data['Language'] == 'en'][['Email','Purchase Price']]
```

```python
# Multiple filtering
sales_data[(sales_data['Job']=='Dancer') & (sales_data['Language'] != 'en')].head()
```

```python
# Filter with column's substring
sales_data[sales_data['CC Exp Date'].apply(lambda s:s[3:] >= '24')].head()
```

```python
sales_data[sales_data['CC Exp Date'].apply(lambda s:s[3:] < '24')].count()
```

```python
# Value Count with column's substring
sales_data['Email'].apply(lambda provider: provider.split('@')[1]).value_counts().head()
```
